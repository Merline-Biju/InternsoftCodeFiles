# Plotting in Pandas

```
a = pd.Series(np.random.normal(size = 50))
print(a)
a.plot()
```

```
a.cumsum().plot(grid = True)
```

```
b = pd.DataFrame({'normal': np.random.normal(size = 10), 
                       'gamma': np.random.gamma(1, size = 10), 
                       'poisson': np.random.poisson(size = 10)})
print(b)

b.cumsum(0).plot()
```

```
b.cumsum(0).plot(subplots = True)
```

```
b.cumsum(0).plot(secondary_y = 'normal')
```

```
fig, axes = plt.subplots(nrows = 3, ncols = 1, figsize = (6, 10))
for i, var in enumerate(['normal','gamma','poisson']):
    b[var].cumsum(0).plot(ax = axes[i], title = var)
    axes[i].set_ylabel('cumulative sum : ' + str(i + 1))
```

# Simple Line Plots

```
fig = plt.figure()
ax = plt.axes()

s = np.linspace(0, 20, 50)
print(s)
ax.plot(s, np.sin(s));
```

```
fig = plt.figure()
ax = plt.axes()

s = np.linspace(0, 20, 50)
print(s)
ax.plot(s, np.cos(s));
```

```
fig = plt.figure()
ax = plt.axes()

s = np.linspace(0, 20, 50)
print(s)
ax.plot(s, np.tan(s));
```

```
plt.plot(s, np.sin(s))
```

***Line Colors and Styles***

```
plt.plot(s, np.sin(s - 0), color='red')   #red     
plt.plot(s, np.sin(s - 1), color='y')     #yellow      
plt.plot(s, np.sin(s - 2), color='0.55')  #greyscale
```

```
plt.plot(s, np.sin(s), color='orange')        
plt.plot(s, np.cos(s), color='g')           
plt.plot(s, np.tan(s), color='0.30') 
```

```
plt.plot(s, np.sin(s), color='orange', linestyle='-')        
plt.plot(s, np.cos(s), color='g', linestyle=':')           
plt.plot(s, np.tan(s), color='0.30', linestyle='-.')
```

# Adjusting the Plot: Axes Limits

```
plt.plot(s, np.sin(s))

plt.xlim(-1, 15)
plt.ylim(-2, 2);
```

# Labeling Plots

```
plt.plot(s, np.cos(s), color='green')
plt.title("Cosine Curve")
plt.xlabel("x ->")
plt.ylabel("cos(x) ->")
```

# Errorbars

```
x = np.linspace(0, 20, 100)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(100)

plt.errorbar(x, y, yerr=dy, fmt='x', color='red');
```

```
x = np.linspace(0, 20, 100)
dy = 0.08
y = np.cos(x) + dy * np.random.randn(100)

plt.errorbar(x, y, xerr=dy, yerr=dy, fmt='d', color='orange');
```

```
data = np.random.randn(10000)
print(data)
plt.hist(data)
```

```
x = np.random.normal(0, 0.8, 500)
y = np.random.normal(-2, 1, 1000)

kwargs = dict(histtype = 'stepfilled', alpha = 0.7, density = True, bins = 50)

plt.hist(x, **kwargs)
plt.hist(y, **kwargs)
```

```
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```

# Two-Dimensional Histograms

```
mean = [5, 2]
cov = [[3, 1], [2, 0.5]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```
plt.hist2d(x, y, bins=25, cmap='Blues')
cb = plt.colorbar()
```

```
plt.hexbin(x, y, gridsize=20, cmap='Purples')
cb = plt.colorbar()
```
