```
def is_prime(num):
    for n in range(2,num):
        if num % n == 0:
            print(str(num) + ' is not prime')
            break
    else: # If never mod zero, then prime
        print(str(num) + ' is prime')

is_prime(6)
is_prime(17)
```

```
import math

def is_prime(num):
    if num % 2 == 0 and num > 2: 
        return False
    for i in range(3, int(math.sqrt(num)) + 1, 2):
        if num % i == 0:
            return False
    return True
```

```
num = int(input('Enter a number : '))

if is_prime(num):
  print( str(num) + " is prime")
else:
  print( str(num) + " is not prime")
```

# **ITERATORS AND GENERATORS**

```
def gencubes(n):
    for num in range(n):
        yield num**3
```

```
for x in gencubes(8):
    print(x)
```

```
def genfibonacci(n):
    a = 1
    b = 1
    for i in range(n):
        yield a
        a,b = b,a+b
```

```
for num in genfibonacci(14):
    print(num)
```

```
def simple_gen():
    for x in range(3):
        yield x
```

```
g = simple_gen()
```

```
print(next(g))
print(next(g))
print(next(g))
#print(next(g)) #Throws Error
```

# **map()**

```
def fahrenheit(T):
    return ((float(9)/5)*T + 32)
def celsius(T):
    return (float(5)/9)*(T-32)
    
temp = [0, 22.5, 40,100]
```

```
F_temps = list(map(fahrenheit, temp))
C_temps = list(map(celsius, F_temps))
print(F_temps)
print(C_temps)
```

```
list(map(lambda x : float(5)/9 * ( x - 32), F_temps))
```

```
a = [11,22,33,44]
b = [55,66,77,88]

list(map(lambda x, y : x+y, a, b))
```
